```python
import numpy as np
import pandas as pd
from sklearn import datasets
from io import StringIO
```

```python
csv_data = '''A,B,C,D,E
5.0,2.0,3.0,,6
1.0,6.0,,8.0,5
0.0,1.0,12.0,4.0,5
3.0,,3.0,5.0,
5.0,1.0,4.0,2.0,4
'''
```

```python
df = pd.read_csv(StringIO(csv_data))
```

```python
df
```

```python
#有任一個missing就排掉
df.dropna()
```

```python
df.dropna(how='all')#全部missing才排
```

```python
#C欄missing才排
df.dropna(subset=['C'])
```

```python
#missing補0
df1=df.fillna(0)
df1
```

```python
#針對B欄missing補B欄平均數
df['B']=df['B'].fillna(df['B'].mean())
df
```

```python
#針對E欄missing補E欄眾數_還能補最小值、最大值等等
df['E'] = df['E'].fillna(df['E'].mode())
df
#df這張表以上被上一個cell改寫過一次
```
